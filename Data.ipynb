```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
insp_data = pd.read_csv("element_level_inspections.csv")
construction_history = pd.read_csv("construction_history.csv")
```

```python
def get_year_str(date_end):
    if int(date_end) > 90:
        return "19" + date_end
    else:
        return "20" + date_end
    
insp_data['ElemInspDate'] = insp_data['ElemInspDate'].map(lambda x: get_year_str(x[6:8]))
```

```python
filtered_data = {}
elem_keys = insp_data['ElementKey'].unique()
for envir_num in range(1, 5):
    envir_data = insp_data[insp_data['EnvironmentKey'] == envir_num]
    for elem_key in elem_keys:
        key = str(envir_num) + "." + str(elem_key)
        filtered_data[key] = envir_data[envir_data['ElementKey'] == elem_key]
```

```python
def sn_normalize(data):
    sn_freq = data['SN'].value_counts()
    sn_nums = sorted(data['SN'].unique())
    sn_norm = np.zeros(len(sn_nums), dtype=np.int32)
    counter = 0
    acc = 0
    for i in sn_nums:
        val = sn_freq[i]
        sn_norm[counter] += val + acc - 1
        acc += val
        counter += 1
    return sn_norm

print(sn_normalize(filtered_data['1.8323']))

def yearConverter(data):
    sn_norm = sn_normalize(data)
    counter = 0
    for i in sn_norm:
        end_next = (i + 1).item()
        min_val = int(data['ElemInspDate'][counter:end_next].astype(str).min())
        data['ElemInspDate'][counter:end_next] = data['ElemInspDate'][counter:end_next].astype(int) - min_val
        counter = end_next
    return data

print(yearConverter(filtered_data['1.8323']))
```

```python
def year_normalize(data):
    year_freq = data['ElemInspDate'].value_counts()
    year_nums = sorted(data['ElemInspDate'].unique())
    year_norm = np.zeros(len(year_nums), dtype=np.int32)
    counter = 0
    acc = 0
    for i in year_nums:
        val = year_freq[i]
        year_norm[counter] += val + acc - 1
        acc += val
        counter += 1
    return year_norm

print(year_normalize(filtered_data['1.8323']))

def get_prob_matrix(data):
#     max_ = int(data['ElemInspDate'].astype(str).max())
    data.sort_values("ElemInspDate", inplace=True)
    counter_list = year_normalize(data)
    matrix = np.zeros(shape=(len(counter_list), 4))

    
#     for i in range(max_ + 1):
#         for j in range(4):
#             elem_perc = data['ElemPercent' + str(j + 1)]
#             if i == 0:
#                 matrix[i][j] = elem_perc[:counter_list[i] + 1].mean() / 100
#             else:
#                 print(i)
#                 matrix[i][j] = elem_perc[counter_list[i - 1] + 1:counter_list[i] + 1].mean() / 100
    runner = 0
    counter = 0
    for i in counter_list:
        end_next = (i + 1).item()
        for j in range(4):
            matrix[counter][j] = data['ElemPercent' + str(j + 1)][runner:end_next].mean()
        runner = end_next
        counter += 1
    return matrix / 100
```

```python
m1 = get_prob_matrix(yearConverter(filtered_data['1.8323']))
print(m1)
print()
m2 = get_prob_matrix(yearConverter(filtered_data['2.8323']))
print(m2)
print()
m3 = get_prob_matrix(yearConverter(filtered_data['3.8323']))
print(m3)
print()
m4 = get_prob_matrix(yearConverter(filtered_data['4.8323']))
print(m4)
print()
```
